```python
import random
import string
import time

# Generate sample products
def generate_products(n):
    products = []
    for i in range(n):
        product_id = f'PRD{i:05}'
        name = ''.join(random.choices(string.ascii_letters, k=10))
        price = round(random.uniform(1.0, 500.0), 2)
        rating = round(random.uniform(1.0, 5.0), 1)
        products.append({
            'product_id': product_id,
            'name': name,
            'price': price,
            'rating': rating
        })
    return products

# Merge Sort Implementation
def merge_sort(products, key):
    if len(products) <= 1:
        return products
    mid = len(products) // 2
    left = merge_sort(products[:mid], key)
    right = merge_sort(products[mid:], key)
    return merge(left, right, key)

def merge(left, right, key):
    result = []
    while left and right:
        if left[0][key] <= right[0][key]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    result.extend(left or right)
    return result

# Hash Table using Python dict
def build_hash_table(products):
    return {p['product_id']: p for p in products}

def search_product(hash_table, product_id):
    return hash_table.get(product_id, "Product not found")

# Test Runner
def run_tests():
    print("=== Running Product Catalogue System Tests ===\n")
    
    # Test sizes
    small_set = generate_products(5)
    medium_set = generate_products(50)
    large_set = generate_products(100000)
    
    ### Test 1: Sort by price (5 items)
    print("Test 1: Sort by price (5 items)")
    start = time.time()
    sorted_small = merge_sort(small_set.copy(), 'price')
    duration = time.time() - start
    print(f"Time taken: {duration:.6f} seconds")
    print(f"First item (lowest price): {sorted_small[0]}\n")
    
    ### Test 2: Sort by rating (50 items)
    print("Test 2: Sort by rating (50 items)")
    start = time.time()
    sorted_medium = merge_sort(medium_set.copy(), 'rating')
    duration = time.time() - start
    print(f"Time taken: {duration:.6f} seconds")
    print(f"First item (lowest rating): {sorted_medium[0]}\n")
    
    ### Test 3: Sort by name (100,000 items)
    print("Test 3: Sort by name (100,000 items)")
    start = time.time()
    sorted_large = merge_sort(large_set.copy(), 'name')
    duration = time.time() - start
    print(f"Time taken: {duration:.3f} seconds")
    print(f"First 3 items: {[p['name'] for p in sorted_large[:3]]}\n")
    
    ### Test 4: Search existing product in hash table
    print("Test 4: Search for existing product (hash table)")
    hash_table = build_hash_table(large_set)
    key = large_set[5000]['product_id']
    start = time.time()
    result = search_product(hash_table, key)
    duration = time.time() - start
    print(f"Time taken: {duration:.8f} seconds")
    print(f"Found: {result}\n")
    
    ### Test 5: Search non-existing product
    print("Test 5: Search for non-existing product")
    start = time.time()
    result = search_product(hash_table, 'NON_EXISTENT')
    duration = time.time() - start
    print(f"Time taken: {duration:.8f} seconds")
    print(f"Result: {result}\n")
    
    ### Test 6: Batch search of 10,000 products
    print("Test 6: Batch search for 10,000 products")
    keys = [large_set[i]['product_id'] for i in range(10000)]
    start = time.time()
    found = [search_product(hash_table, k) for k in keys]
    duration = time.time() - start
    print(f"Time taken: {duration:.3f} seconds for 10,000 lookups\n")
    
    ### Test 7: Sort and search combined (hybrid test)
    print("Test 7: Sort and search (hybrid system)")
    start = time.time()
    _ = merge_sort(large_set.copy(), 'rating')
    _ = search_product(hash_table, large_set[7777]['product_id'])
    duration = time.time() - start
    print(f"Time taken (combined): {duration:.3f} seconds\n")
    
    print("=== All Tests Completed ===")

# Run the test suite
if __name__ == "__main__":
    run_tests()
```
